# Matplotlib

Matplotlib is a python graph plotting library helps us to plot our data in form of some visualization or figures 

## Why matplotlib
- built on numpy arrays and python
- integrates directly with pandas
- can create basic or advance plots 
- simple to use and provides large number of built-in functions

## Matplotlib example workflow

```python
# 1. import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# 2. prepare some data
x = [1,2,3,4]
y = [11,17,13,20]

# 3. Setup plot
fig , ax = plt.subplots(figsize=(10,10))      #(width, height)


# 4. plot the data
ax.plot(x,y)

# 5. Customize plot
ax.set(title="Simple Plot",
      xlabel = "x-axis",
      ylabel="y-axis")

# 6. save and show
fig.savefig('images/sample-plot.png')
```

```python
arr1 = np.array([45,8,96,45,2])
arr2 = np.array([-89,10,10,20,8])
```

```python
arr1, arr2
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almod butter": 10,
                    "Peanut butter": 8,
                    "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title="Nut Butter store",
      xlabel = "Butter type" ,
      ylabel="Rate per doller");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
```

```python
x = np.random.randn(1000)
x
```

```python
# createing a histogram

fig, ax = plt.subplots()
ax.hist(x);
```

```python
## Creating a subplot
fig,ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10,10))
ax[0,0].plot(x,x**2)
ax[0,1].scatter(x,x*2)
ax[1,0].bar(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
ax[1,1].hist(np.random.randn(200));
```

## Which one should you use? (pyplot vs OO method)

* When plotting something quickly it's ok to use pyplot method
* But if plotting something complex then we can go with OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
```

```python
over_50
```

```python
over_50.reset_index(drop=True, inplace=True)
```

```python
over_50
```

```python
# pyplot method
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            figsize=(10,6),
            xlabel='age');
```

```python
# OO method

fig ,ax = plt.subplots(figsize =(10,6))
over_50.plot(kind="scatter",
            x ="age",
            y="chol",
            c="target",
            ax=ax);
# ax.set_xlim([45,100])
```
